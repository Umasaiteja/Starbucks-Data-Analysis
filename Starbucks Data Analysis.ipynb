```python
caffeine_calories_df = df[['Beverage', 'Caffeine (mg)', 'Calories']]
print(caffeine_calories_df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a heatmap to visualize the relationship between caffeine content and calorie count
plt.figure(figsize=(10, 8), facecolor='white')
heatmap_data=df.pivot_table(index='Beverage', columns='Beverage_prep', values=['Caffeine (mg)', 'Calories'], aggfunc='mean')
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('Heatmap of Caffeine Content and Calorie Count in Starbucks Beverages')
plt.xlabel('Beverage Preparation')
plt.ylabel('Beverage')
plt.show()
```

```python
specific_beverages = ['Caffè Latte', 'Caffè Mocha', 'Cappuccino', 'Espresso']
filtered_df = df[df['Beverage'].isin(specific_beverages)][['Beverage', 'Beverage_prep', 'Caffeine (mg)', 'Calories']]
print(filtered_df)
```

```python
plt.figure(figsize=(10, 6), facecolor='white')
sns.scatterplot(data=df, x='Caffeine (mg)', y='Calories', hue='Beverage_prep', style='Beverage_prep', s=100)
plt.title('Scatter Plot of Caffeine Content vs Calories')
plt.xlabel('Caffeine (mg)')
plt.ylabel('Calories')
plt.legend(title='Beverage Preparation', bbox_to_anchor=(1.05, 1), loc=2)
plt.grid(True)
plt.show()
```

```python
df['Caffeine (mg)'] = pd.to_numeric(df['Caffeine (mg)'], errors='coerce')
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
```

```python
sns.violinplot(data=df, x='Caffeine (mg)', y='Calories', hue='Beverage_prep', style='Beverage_prep', s=100)
plt.title('Caffeine and calories')
plt.show()
```

```python
x = df['Caffeine (mg)']
y = df['Calories'] 
plt.stem(x, y, markerfmt='o', linefmt='b-', label='Beverage Data')
plt.title('Caffeine and calories ')
plt.xlabel('Caffeine (mg)')
plt.ylabel('Calories')
plt.show()
```
